```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("..\..\cmake-build-debug-visual-studio\output.csv")
```

```python

fig, ax = plt.subplots(2, 2, figsize=(15, 15))
for i, sort_type in enumerate(df['type'].unique()):
    sub_data = df[df['type'] == sort_type]
    axes = sns.lineplot(
        x=sub_data['size'], 
        y=sub_data['time'], 
        hue=sub_data['sort'], 
        ax=ax[i // 2, i % 2],
    )
    axes.set_title(sort_type)
    
plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(16, 8))
for i, sort_type in enumerate(df['sort'].unique()):
    sub_data = df[df['sort'] == sort_type]
    axes = sns.lineplot(
        x=sub_data['size'],
        y=sub_data['time'],
        hue=sub_data['type'],
        ax=ax[i],
    )
    axes.set_title(sort_type)

plt.show()
```
